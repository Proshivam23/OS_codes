```python
from collections import deque

def equal_tuple(tup1, tup2, n):
    for i in range(n):
        if tup1[i] != tup2[i]:
            return False
    return True

def is_valid(initialState, currentState):
    m1 = initialState[0]
    m2 = currentState[0]
    cl = initialState[1]
    c2 = currentState[1]
    if m2 < c2 and m2 != 0:
        return False
    if m2 != 0 and (cl - c2) > (m1 - m2) and (m1 - m2 != 0) and m2 != 0:
        return False
    return True

def successorStates(initialState, givenState):
    m1 = initialState[0]
    cl = givenState[1]
    boat = 1 - givenState[2]
    states = []
    extraStates = []

    if boat == 1:
        print("we are going to the right")
        try:
            for i in range(0, m1+1):
                for j in range(0, cl+1):
                    if (i + j) > 0 and (i + j) <= 2:
                        if (m1 - i, cl - j, boat) not in (list(visited.values()) + list(visited.keys())):
                            if (m1 - i) >= 0 and (cl - j) >= 0:
                                if (m1 - i == 0) and (cl - j == 0):
                                    print("Reached success")
                                    visited[(0, 0, 1)] = givenState
                                    extraStates.append((0, 0, 1))
                                    return extraStates
                                if is_valid(initialState, (m1 - i, cl - j, boat)):
                                    extraStates.append((m1 - i, cl - j, boat))
        except Exception as e:
            print(f"cMiss={m1-i}, cCann={cl-j}")

        for state in extraStates:
            if is_valid((m1, cl, 0), state):
                visited[state] = givenState
                states.append(state)

    elif boat == 0:
        print("we are going to the left")
        extraStates = []
        try:
            for i in range(0, m1+1):
                for j in range(0, cl+1):
                    if (i + j) > 0 and (i + j) <= 2:
                        if (m1 + i, cl + j, boat) not in (list(visited.values()) + list(visited.keys())):
                            if (m1 + i) <= m1 and (cl + j) <= cl:
                                if is_valid(initialState, (m1 + i, cl + j, boat)):
                                    extraStates.append((m1 + i, cl + j, boat))
        except IndexError as e:
            print(f"cMiss={givenState[0]}, cCann={givenState[1]}")

        for state in extraStates:
            if is_valid((m1, cl, 0), state):
                visited[state] = givenState
                states.append(state)

    return states

def mnc(initialState):
    q = deque()
    q.append(initialState)
    j = 0

    while q:
        c = q.popleft()
        print(f"current={c}")
        j += 1

        if equal_tuple(c, (0, 0, 1), 3):
            print(f"found a solution: {c}")
            solutionQ = list()
            x = (0, 0, 0)
            s = successorStates(initialState, c)
            solutionQ.append((0, 0, 1))
            t = (0, 0, 1)
            
            while x != initialState:
                x = visited[t]
                solutionQ.append(x)
                t = x
            
            print("Solution Q:")
            for i in solutionQ[::-1]:
                print(f"=> {i} ", end="")
            return

        else:
            successor = successorStates(initialState, c)
            if not successor:
                print("No new states, kill this branch")
            else:
                print(f"New states are {successor}")
                for state in successor:
                    q.append(state)

# Initialize variables
visited = {(): ()}  # where each child stores its parent

# Example usage
initialState = (3, 3, 0)
mnc(initialState)
```
