```python
from collections import deque

def equal_tuple(tup1, tup2, n):
    for i in range(n):
        if tup1[i] != tup2[i]:
            return False
    return True

def is_valid(initial_state, current_state):
    m1, c1, b1 = initial_state
    m2, c2, b2 = current_state

    if m2 < c2 and m2 != 0:
        return False

    if m2 != 0 and (c1 - c2) > (m1 - m2) and (m1 - m2 != 0) and m2 != 0:
        return False

    return True

def successor_states(initial_state, given_state, visited):
    st = []  # contains the successful solution lists, and therefore, is supposed to be a list of lists
    ml, cl, _ = given_state
    m2, c2, _ = initial_state
    boat = 1 - given_state[2]  # we are changing the side at the start itself
    states = []

    if boat == 1:
        print("we are going to the right")
        extra_states = []
        for i in range(0, m2):
            for j in range(0, c2):
                if (i + j) > 0 and (i + j) <= 2:
                    if (ml - i, cl - j, boat) not in (list(visited.values()) + list(visited.keys())):
                        if (ml - i) >= 0 and (cl - j) >= 0:
                            if (ml - i) == 0 and (cl - j) == 0:
                                print("Reached success")
                                visited[(0, 0, 1)] = given_state
                            if is_valid(initial_state, (ml - i, cl - j, boat)):
                                extra_states.append((ml - i, cl - j, boat))
        for state in extra_states:
            if is_valid(initial_state, state):
                visited[state] = given_state
                states.append(state)
    elif boat == 0:
        print("we are going to the left")
        extra_states = []
        for i in range(0, m2):
            for j in range(0, c2):
                if (i + j) > 0 and (i + j) <= 2:
                    if (ml + i, cl + j, boat) not in (list(visited.values()) + list(visited.keys())):
                        if (ml + i) <= m2 and (cl + j) <= c2:
                            if is_valid(initial_state, (ml + i, cl + j, boat)):
                                extra_states.append((ml + i, cl + j, boat))
        for state in extra_states:
            if is_valid(initial_state, state):
                visited[state] = given_state
                states.append(state)

    return states

def mnc(initial_state):
    # takes a 3 item list as input, first is missionaries, second is cannibals, third is boat(0 on left, 1 on right)
    # initialise a queue
    q = deque()
    q.append(initial_state)
    # assuming initial state is a 3-item tuple
    # now the bfs loop
    i = 0
    visited = {}
    visited[(0, 0, 0)] = (0, 0, 0)

    while q:
        c = q.popleft()
        print(f"Current state: {c}")
        i += 1

        if equal_tuple(c, (0, 0, 1), 3):
            print(f"Found a solution: {c}")
            solution_q = []
            x = (0, 0, 0)
            t = (0, 0, 1)
            solution_q.append((0, 0, 1))

            while x != initial_state:
                x = visited[t]
                solution_q.append(x)
                t = x

            print("Solution Q:")
            for i in solution_q[::-1]:
                print(f"=>{i}", end="")

        else:
            successors = successor_states(initial_state, c, visited)

            if not successors:
                print("No new states, kill this branch")
            else:
                print(f"New states are {successors}")
                for state in successors:
                    q.append(state)

# Call the function with the initial state
mnc((3, 3, 0))
```
